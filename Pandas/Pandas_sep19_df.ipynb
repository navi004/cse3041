### Group By

```python
#group by newcolumn and get the sum
df.groupby('newcolumn').sum()
```

```python
#max groupby newcolumn
df.groupby('newcolumn').max()
```

```python
#Getting mean of avg_low group by newcolumn
df.groupby('newcolumn')['avg_low'].std()

# returns the mean of avg_low grouped by newcolumn, which means the avg_low column is grouped in 0's , 1's , 2's 
# then gets the average of each group 
```

```python
df.groupby('newcolumn').aggregate({'avg_low':["min","max"], 'avg_high':["mean","median"]})
```

```python
df.groupby('newcolumn').aggregate({'avg_low':[min,max], 'avg_high':[np.mean,np.median]})
```
