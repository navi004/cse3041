# Pandas DataFrame

```python
import pandas as pd
```

**A dataframe is a two dimensional data structure, i.e, data is aligned in a tabular fashion in rows and columns**

* Potentially columns are different data types

* Size -- Mutables

* Labeled axes (rows and columns)

* Can Perform Arithmetic operations on rows and columns

* **Creating an Empty Data Frame**

```python
#Creating an Empty DataFrame
df = pd.DataFrame()
print(df)
```

* **Append Columns to an Empty DataFrame**

```python
#Append Columns to an empty DataFrame
df['Name'] = ['Ankir','Anitha','Yash']
df['Age'] = [19,18,20]
df['Gender'] = ['M','F','M']
df
```

* **Delete Age from df**

```python
#Delete age 
del(df['Age'])
```

```python
df
```

### **Creating from Dictionary**

```python
data = {'sub':['Maths','Science','Social Science','Computer Science'],'marks':[100,98,87,89],'grade':['A','A','B','B']}
data
df = pd.DataFrame(data)
```

```python
df
```

```python
pd.Series(data)
```

### Creating DataFrame from Series

```python
subj = pd.Series(['Maths','Science','Social Studies','Computer Science'])
marks = pd.Series([100,98,87,89])
grade = pd.Series(['A','A','B','B'])

#df = pd.DataFrame([subj,marks,grade])
#df

df = pd.DataFrame([subj,marks,grade],index = ['subj','marks','grade']).T
df
```

#### While converting series into DataFrame need to transpose before displaying the output
