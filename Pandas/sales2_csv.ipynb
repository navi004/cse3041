```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python

#Read sales2.csv file into a dataframe
df1 = pd.read_csv("/content/sales2.csv")
df1
```

```python
df1[df1['customer_id'] == 3001]
```

```python
#Update the purchase amount of order Id 70012 as 500.
df1.loc[df1['ord_no'] == 70012,'purch_amt'] = 500
df1
```

```python
#Retrieve the order no and customer id of the maximum purchase amount
df1.loc[(df1.purch_amt == max(df1.purch_amt)), ['ord_no', 'customer_id', 'purch_amt']]
```

```python
#Retrieve the order no, customer id and purchase amount whose value is greater than
1000.
df1.loc[(df1.purch_amt > 1000), ['ord_no', 'customer_id', 'purch_amt']]
```

```python
#Fetch the order details that are placed during the month of October.
df1['ord_date'] = pd.to_datetime(df1['ord_date'])
df1
```

```python
df1[df1['ord_date'].dt.month == 10]
```

```python
# Split the dataset into groups on customer_id to summarize purch_amt with respect to
#each salesman id
g = df1.groupby('customer_id')
for x in g:
  print(x)
```

```python
gr_data = df1.groupby(['customer_id', 'salesman_id']).agg({'purch_amt':'sum'})
gr_data
```

```python
sum(gr_data.iloc[:,0])
```

```python
#Calculate percentage of purch_amt in each group and rename purch_amt to
#purchaseAmount.
gr_data['percentage'] = gr_data.iloc[:,0] * 100/sum(gr_data.iloc[:,0])
gr_data['percentage'] = round(gr_data['percentage'],2)
gr_data
```
