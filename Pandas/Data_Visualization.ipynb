# Pandas Data Visualization

Data Visualization is a technique used for the graphical representation of data. (Eg.: scatterplot, line, bar, piechart, histrogram).

# **Libraries**
*    **Pandas** :tabular data manipulation
*    **matplot lib** : simple plots like lines, scatterplot etc..
*    **seaborn** : more complex plots like correlation etc.
*    **ggplot2** : Grammer of graphics, can be used along with R




```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/plotdata.csv')
df
```

# Scatter Plot

```python
df.plot(x = 'Unemployment_rate', y = 'Index_price', kind = 'scatter')
plt.title("Scatter PLot")
plt.show()
```

# Bar Chart

```python
df.plot(x = 'Year', y = 'Unemployment_rate', kind = 'bar')
plt.title('Bar Chart')
plt.show()
```

# Line Chart

```python
df.plot(x = 'Year', y = 'Unemployment_rate', kind = 'line')
plt.title('Line Chart')
plt.show()
```

# Pie Chart

```python
df.plot(x = 'Unemployment_rate', y = 'Year', kind = 'pie')
plt.title('Pie Chart')
plt.show()
```

# Histogram

```python
df.plot(x = 'Index_price', y = 'Unemployment_rate', kind = 'hist')
plt.title('Histogram')
plt.show()
```
